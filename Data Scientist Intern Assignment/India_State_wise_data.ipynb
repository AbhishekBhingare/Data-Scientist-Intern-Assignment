```python
# Q9 SHOW TYPES OF WORKER AND WHICH HAS MORE AND LESS POPULATION

Male_Workers = df['Male_Workers'].sum()
Female_Workers = df['Female_Workers'].sum()
Main_Workers = df['Main_Workers'].sum()
Marginal_Workers = df['Marginal_Workers'].sum()
Non_Workers = df['Non_Workers'].sum()
Cultivator_Workers = df['Cultivator_Workers'].sum()
Agricultural_Workers = df['Agricultural_Workers'].sum()
Household_Workers = df['Household_Workers'].sum()
Other_Workers = df['Other_Workers'].sum()

total_workers = [Male_Workers, Female_Workers, Main_Workers,Marginal_Workers, Non_Workers, Cultivator_Workers, Agricultural_Workers, Household_Workers, Other_Workers ]
lables = ['Male_Workers', 'Female_Workers', 'Main_Workers','Marginal_Workers', 'Non_Workers', 'Cultivator_Workers', 'Agricultural_Workers', 'Household_Workers','Other_Workers' ]
plt.figure(figsize=(20,6))
plt.xlabel('Workers')
plt.ylabel('No of Workers')
plt.title('Tyes of workers')
plt.bar(lables, total_workers)
```
